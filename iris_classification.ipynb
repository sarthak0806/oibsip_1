Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

Dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.describe()
```

Visualisation


```python

sns.pairplot(df,hue='Species')
```

Seperate feature and target(slicing of matrix)

```python
data = df.values
X = data[:,1:5]
Y = data[:,5]
print(X)
print(Y)
```

Splitting the data into training and testing for further model development

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
print(X_test)
```

Logistic Regression

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
prediction = model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
accu = int(accuracy_score(y_test,prediction)*100)
print(f"Our model accuracy is {accu}%.")
for i in range(len(prediction)):
  print(y_test[i],"          ",prediction[i])
```
